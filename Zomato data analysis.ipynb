# Zomato Data Analysis Porject

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataframe = pd.read_csv("Zomato data .csv")
print(dataframe)
```

```python
dataframe
```

# Convert the data type of column rate 

```python
def handleRate(value):
    value = str(value).split('/')
    value=value[0];
    return float(value)

dataframe['rate'] = dataframe['rate'].apply(handleRate)
print(dataframe.head())
```

```python
dataframe.info()
```

# Type of resturant

```python
dataframe.head()
```

```python
sns.countplot(x=dataframe['listed_in(type)'])

# Correctly set the x-axis label
plt.xlabel("Type of Restaurant")

# Display the plot
plt.show()
```

```python
dataframe.head()
```

```python
grouped_data=dataframe.groupby('listed_in(type)')['votes'].sum()
result=pd.DataFrame({'votes':grouped_data})
plt.plot(result, c="green", marker="o")  # Corrected 'marker' keyword
plt.xlabel("Type of Restaurant", c="blue", size=20)
plt.ylabel("Votes", c="red", size=20)

# Display the plot
plt.show()
```

# conclusion- dinning restaurant has revieved maximum votes

```python
plt.hist(dataframe['rate'],bins=10)
plt.title("ratings distributions")
plt.show()
```

# Average order spending by couples

```python
dataframe.head()
```

```python
cost_data = dataframe['approx_cost(for two people)']

# Plotting the countplot
sns.countplot(x=cost_data)

# Setting labels and title
plt.xlabel('Approximate Cost (for two people)')
plt.ylabel('Count')
plt.title('Cost Distribution for Two People')
plt.show()
```

```python
pivot_table = dataframe.pivot_table(index='listed_in(type)', columns='online_order', aggfunc='size', fill_value=0)

# Plotting the heatmap
sns.heatmap(pivot_table, annot=True, cmap="YlGnBu", fmt='d')  # Note: 'Y1GnBu' was corrected to 'YlGnBu'
plt.title("Heatmap")
plt.xlabel("Online Order")
plt.ylabel("Listed In (Type)")
plt.show()
```
